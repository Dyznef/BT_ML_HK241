```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
import pandas as pd
```

```python
x = np.array([155, 180, 164, 162, 181, 182, 173, 190, 171, 170, 181, 182, 189, 184, 209, 210])
y = np.array([51, 52, 54, 53, 55, 59, 61, 59, 63, 76, 64, 66, 69, 72, 70, 80])

theta0 = 0
theta1 = 0
alpha = 0.00001
iterations = 1000
```

```python
# h function
def h(x, theta0, theta1):
    return theta0 + theta1 * x

# Cost function
def cost(X, Y, theta0, theta1):
    m = len(X)
    return 1/(2*m) * np.sum((h(X, theta0, theta1) - Y)**2)

# Gradient descent
def gradient_descent(X, Y, theta0, theta1, alpha, iterations):
    m = len(X)
    
    for i in range(iterations):
        grad0 = (1/m) * np.sum(h(X, theta0, theta1) - Y)
        grad1 = (1/m) * np.sum((h(X, theta0, theta1) - Y) * X)       
        theta0 -= alpha * grad0
        theta1 -= alpha * grad1

    return theta0, theta1
```

```python
theta0, theta1 = gradient_descent(x, y, theta0, theta1, alpha, iterations)

plt.scatter(x, y, color='red', label='Data')
plt.plot(x, h(x, theta0, theta1), color='blue', label=f'theta(x) = {theta0:.6f} + {theta1:.6f}x')

plt.xlabel('Thời gian học (h)')
plt.ylabel('Điểm thi')
plt.title('Đồ thị hồi quy tuyến tính')
plt.legend()

plt.show()
```
